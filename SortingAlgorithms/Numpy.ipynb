```python
import numpy as np

# 1D array
arr = np.array([1, 2, 3, 4])
print(arr)

# 2D array
arr_2d = np.array([[1, 2], [3, 4]])
print(arr_2d)
```

```python
# Array of zeros
zeros = np.zeros((2, 3))  # 2x3 matrix of zeros

# Array of ones
ones = np.ones((3, 3))  # 3x3 matrix of ones

# Array with a range of values
arange = np.arange(1, 10, 2)  # Start: 1, Stop: 10, Step: 2

# Linearly spaced values
linspace = np.linspace(0, 1, 5)  # 5 equally spaced values between 0 and 1

print(zeros,ones,arange,linspace)
```

```python
# Random values
random = np.random.rand(3, 3)  # 3x3 matrix of random values (0 to 1)

# Random integers
randint = np.random.randint(1, 10, size=(3, 3))  # Random integers between 1 and 10
print(randint)
random
```

```python
arr = np.array([1, 2, 3, 4, 5])

# Min, max, sum, mean, standard deviation
print(np.min(arr))
print(np.max(arr))
print(np.sum(arr))
print(np.mean(arr))
print(np.std(arr))
```

```python
arr = np.arange(1, 10)

# Reshape to 3x3 matrix
reshaped = arr.reshape((3, 3))
print(reshaped)
```
